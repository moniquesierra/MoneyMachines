```python
# In this assignment you will read a couple of Stock indexes from the Internet and will plot some of their information.

import pandas as pd
from pandas_datareader import data, wb
import numpy as np
import datetime
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Read the stock information for Amazon (Tick: AMZN) using Pandas’ data reader

start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)

# AMZN

AMZN = data.DataReader("AMZN" , "yahoo", start, end)
AMZN
```

```python
# Make sure to use the ‘ggplot’ style for all Amazon’s charts
# Plot a time series (line chart) of the “volume” from January 1, 2022 to today.


myAMZN = AMZN.reset_index()
myAMZN

myAMZN.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
plt.show()
```

```python
# Plot a bar chart of the “opening “ for the month of December 2021

dec_start = datetime.datetime(2021, 12, 1)
dec_end = datetime.datetime(2021, 12, 31)


AMZN_dec = data.DataReader("AMZN" , "yahoo", dec_start, dec_end)

AMZN_dec["Open"].plot.bar(figsize=(12,3),lw=1)
plt.show()
```

```python
# Read the stock information for Microsoft (Tick: MSFT) using Pandas’ data reader


start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)

MSFT = data.DataReader("MSFT" , "yahoo", start, end)
MSFT
```

```python
# Plot a time series (line chart) of the “volume” from January 1, 2022 to today.

myMSFT = MSFT.reset_index()
myMSFT

plt.style.use('bmh')
myMSFT.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
plt.show()
```

```python
# Plot a density chart of the ‘High’ and the ‘Low’  values for the month of January 2022

start1 = datetime.datetime(2022, 1, 1)
end1 = datetime.datetime(2022, 1, 31)

myMSFT[["High", "Low"]].plot.density()
plt.show()
```
